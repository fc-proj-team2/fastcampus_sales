## 가장 많이 사용된 쿠폰
- 사람들이 가장 많이 사용한 쿠폰은 어떤 것일까?

- 일반 수강생에게 제공하지 않는 쿠폰, 내부 테스트용 쿠폰은 무엇이 있고 어떻게 처리할까?

```python
coupon_df = pd.DataFrame()

coupon_count = df['쿠폰이름'].value_counts()
coupon_name = df['쿠폰이름'].unique()

cnt = []
for i in range (len(coupon_count)):
    cnt.append(coupon_count[i])
coupon_df['counts'] = cnt
coupon_df['coupon_name'] = df['쿠폰이름'].unique()
coupon_df.head()
```

### 무의미한 쿠폰 삭제 
- 내부 테스터용
- 특정 조건을 수립해야 (학원 수강생, 트위터 팔로우, 강사 지인) 받을 수 있는 쿠폰
- 산학협력 

```python
# 년도를 벗어난 자기계발 어워즈 무료수강권 쿠폰 삭제
# e.g) 2056 자기계발 어워즈 무료수강권 

year_coupon = list(coupon_df[coupon_df['coupon_name'].str.contains('자기계발 어워즈 무료수강권')].index)
year_coupon = year_coupon[1:] # 0번째만 정상 쿠폰
year_coupon
```

```python
# 내부 테스트용 값 확인

stopword = ['^테스트$', '결제테스트', '검수용', '미리보기', '업무용', '참고용', '미리보기', 'test', '검토용', '신해동', '내부', '직원', '지인', '강의검수', '조교', '제공용', '예외처리', 'cx', 'CX', '^쿠폰$', '^3$', '발행', '예외발급', 'TEST', '인턴']

# 'coupon_name' 컬럼에서 stopword 리스트 안에 있는 단어가 포함된 행만 선택
filtered_df = coupon_df[coupon_df['coupon_name'].str.contains('|'.join(stopword), na=False)
                        & (~coupon_df['coupon_name'].str.contains('A/B 테스트'))]
```

```python
# 삭제할 쿠폰 행들만 포함되어 있는 df
filtered_df
```

## 유의미한 쿠폰
- 쿠폰 카테고리 추출
- e.g) [WELCOME]...


### 쿠폰 형식
1. [카테고리]쿠폰이름
2. [재구매/재발급/리워드][카테고리]쿠폰이름
3. 쿠폰이름

```python
# 각 형식을 담을 리스트 생성
coupon_category = [] #[]
double_cate = [] # [][]
non_cate = []

for c in filtered_coupon['coupon_name']:
    # [xxx][xxx]
    # 대괄호 갯수에 따라 함수 호출
    if c.count(']') > 1:
        double_cate.append(c)
    elif c.count(']') == 1:
        coupon_category.append(c)
    else:
        non_cate.append(c)
```

```python
value = []

# 대괄호 안 카테고리 추출
def retrieve_category(arr, cond):
    if cond == 2:
        for i in range (len(arr)):
            val = arr[i].split(']')[0]
            val = val.split('[')[1]
            print(val)
            value.append(val)
    if cond == 1:
        for i in range (len(arr)):
            val = arr[i].split(']')[0]
            if val.count('[') == 1:
                val = val.split('[')[1]
            value.append(val)
    
retrieve_category(coupon_category, 2)
retrieve_category(coupon_category, 1)

categories = pd.DataFrame(value)
```

```python
# 추출한 쿠폰 값을 데이터 프레임으로 만들기

categories = pd.DataFrame(value)
```

#### 카테고리 빈도 수에 따라 횟수 세기

```python
free_lecture = 0
div_pay = 0
target = 0
discount20 = 0
discount50 = 0
discount100 = 0

for v in value:
    # 무료수강권
    if '무료수강권' in v:
        free_lecture += 1
        
    if '대상' in v:
        target += 1
        
    if '전용' in v:
        target += 1
        
    # 분할결제
    if '분할' in v:
        div_pay += 1
    
    # 20% 할인
    if '20' in v:
       discount20 += 1
        
    # 50% 할인
    if '50' in v:
        discount50 += 1
        
    # 100% 할인
    if '100' in v: 
        discount100 += 1   
    
```

```python
print(free_lecture,
div_pay,
target,
discount20,
discount50,
discount100
)
```
